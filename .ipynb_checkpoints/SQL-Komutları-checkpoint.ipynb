	-- AGG KOMUTLARI
SELECT 
	BRANCH AS [SUBE ADI],
	SUM(LINENET) AS [TOPLAM SATIS],
	COUNT(LINENET) AS [SATIS SAYISI],
	MIN(LINENET) AS [MINIMUM SATIS],
	MAX(LINENET) AS [MAKSIMUM SATIS],
	AVG(LINENET) AS [ORTALAMA SATIS]
FROM
	SALES
GROUP BY
	BRANCH
ORDER BY
	BRANCH

	-- GROUP BY VE HAVING KOMUTU
SELECT 
	BRANCH AS [SUBE ADI],
	COUNT(LINENET) AS [SATIS SAYISI]
FROM
	SALES
GROUP BY
	BRANCH
HAVING 
	COUNT(LINENET) > 15000
ORDER BY
	[SATIS SAYISI] DESC

	-- ŞUBELERİN GÜNLÜK SATIŞLARI
SELECT
	DATE_,
	BRANCH,
	SUM(LINENET) AS SATIS
FROM
	SALES
GROUP BY 
	DATE_, BRANCH
ORDER BY
	DATE_ 

	-- KATEGORİLERE GÖRE SATIS TUTARI
SELECT 
	CATEGORY_NAME1, 
	CATEGORY_NAME2,
	SUM(LINENET) AS [SATIS TUTARI]
FROM 
	SALES
WHERE
	CATEGORY_NAME1 IS NOT NULL
GROUP BY 
	CATEGORY_NAME1, 
	CATEGORY_NAME2
HAVING 
	SUM(LINENET) > 30000
ORDER BY
	[SATIS TUTARI] DESC

	-- MUSTERILERIN GITTIGI MAGAZA SAYILARI
SELECT
	CLIENTNAME, 
	COUNT(DISTINCT BRANCH) AS [MAGAZA SAYISI]
FROM
	SALES
WHERE
	CLIENTNAME IS NOT NULL
GROUP BY 
	CLIENTNAME
HAVING 
	COUNT(DISTINCT BRANCH) > 1
ORDER BY
	[MAGAZA SAYISI] DESC

SELECT 
	U.ID AS ID,
	U.USERNAME_ AS [KULLANICI ADI],
	U.NAMESURNAME AS [ADI SOYADI],
	U.EMAIL AS [E-POSTA ADRESİ],
	U.TELNR1 AS [TEL NO],
	COUNT(DISTINCT O.ID) AS [SİPARİŞ SAYISI],
	SUM(DISTINCT BD.TOTAL) AS [TOPLAM TUTAR],
	COUNT(DISTINCT BD.ITEMID) AS [ÜRÜN ÇEŞİDİ]
FROM
	USER_ AS U
JOIN ORDER_ AS O ON U.ID=O.USERID
JOIN BASKET AS B ON U.ID=B.USERID
JOIN BASKETDETAIL AS BD ON B.ID=BD.BASKETID
GROUP BY 
	U.USERNAME_,
	U.NAMESURNAME,
	U.EMAIL,
	U.TELNR1,
	U.ID
ORDER BY 
	[SİPARİŞ SAYISI] DESC,
	[ÜRÜN ÇEŞİDİ] DESC,
	[TOPLAM TUTAR] DESC

SELECT 
	U.ID AS ID,
	U.USERNAME_ AS [KULLANICI ADI],
	U.NAMESURNAME AS [ADI SOYADI],
	C.CITY AS [ŞEHİR],
	COUNT(DISTINCT O.ID) AS [SİPARİŞ SAYISI],
	SUM(DISTINCT BD.TOTAL) AS [TOPLAM TUTAR],
	COUNT(DISTINCT BD.ITEMID) AS [ÜRÜN ÇEŞİDİ]
FROM
	USER_ AS U
JOIN ORDER_ AS O ON U.ID=O.USERID
JOIN BASKET AS B ON U.ID=B.USERID
JOIN BASKETDETAIL AS BD ON B.ID=BD.BASKETID
JOIN ADDRES AS A ON O.ADDRESSID=A.ID
JOIN CITY AS C ON A.CITYID=C.ID
GROUP BY 
	U.ID,
	U.USERNAME_,
	U.NAMESURNAME,
	C.CITY
ORDER BY 
	[KULLANICI ADI]

SELECT 
	C.CITY,
	SUM(CASE WHEN U.GENDER='K' THEN BD.TOTAL ELSE 0 END) AS [TOPLAM-KADIN],
	SUM(CASE WHEN U.GENDER='E' THEN BD.TOTAL ELSE 0 END) AS [TOPLAM-ERKEK]
FROM ORDER_ AS O
JOIN USER_ AS U ON U.ID=O.USERID
JOIN ADDRES AS A ON A.ID=O.ADDRESSID
JOIN CITY AS C ON C.ID=A.CITYID
JOIN BASKET AS B ON B.ID=O.BASKETID
JOIN BASKETDETAIL AS BD ON BD.BASKETID=B.ID
GROUP BY C.CITY
ORDER BY C.CITY
